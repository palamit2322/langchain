## PromptTemplate


```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()
model=ChatOpenAI(model="gpt-4o-mini", temperature=0.5)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template=PromptTemplate.from_template("what is the capital of: {country} ")
prompt=prompt_template.format(country="India")
response=model.invoke(prompt)
print(response.content)
```

## ChatPromptTemplate

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template=ChatPromptTemplate(
    [
        ("system","You are an helpful assistant."),
        ("user","What is the capital of {country}")
    ]
)
prompt=chat_template.format_messages(country="India")

response=model.invoke(prompt)
print(response.content)
```

## MessagePlaceHolder

```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
from langchain_core.messages import HumanMessage

prompt_template=ChatPromptTemplate(
    [
        ("system","You are an helpful assistant."),
        MessagesPlaceholder("usgMsg")
    ]
)
message=[HumanMessage(content="What is the capital of India")]
chain=prompt_template|model
response=chain.invoke({"usgMsg":message})

print(response.content)

```
