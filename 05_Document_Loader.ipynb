## Document Loader

## Text Loader

```python
from langchain_community.document_loaders import TextLoader

path="requirements.txt"

loader=TextLoader(path)
docs=loader.load()

print(len(docs))
print(type(docs))
print(docs[0].page_content)
```

## PyPDFLoader

```python
from langchain_community.document_loaders import PyPDFLoader
from pathlib import Path

BASE_DIR =Path.cwd()
print(BASE_DIR)

PDF_PATH = BASE_DIR / "data" / "MachineLearning.pdf"
print(PDF_PATH)

loader=PyPDFLoader(str(PDF_PATH))
docs=loader.load()

print(len(docs))
print(type(docs))
print(docs[0].page_content)
```

## CSV Loader

```python
from langchain_community.document_loaders import CSVLoader

from pathlib import Path

BASE_DIR=Path.cwd()

CSV_PATH=BASE_DIR / "data" / "ecomm.csv"

loader=CSVLoader(CSV_PATH)
docs=loader.load()

print(len(docs))
print(type(docs))
print(docs[0].page_content)
```
